## Buliding a regression model

The data have already been split to predictor (*data*) and response (*target*) variables. Given this information, we'll apply what we have previously learned about linear regression and apply the algorithm to the diabetes dataset.

#### In the cell below, import the `linear_model` class from `sklearn`. 

```python
from sklearn.linear_model import LinearRegression
```

#### Create a new instance of the linear regression model and assign the new instance to the variable `diabetes_model`.

```python
diabetes_model = LinearRegression()
```

#### Next, let's split the training and test data.

Define `diabetes_data_train`, `diabetes_target_train`, `diabetes_data_test`, and `diabetes_target_test`. Use the last 20 records for the test data and the rest for the training data.

```python
# Forma 1 a mano

X_train, X_test, y_train, y_test  = (diabetes["data"][:-20, :], 
                                     diabetes["data"][-20:, :], 
                                     diabetes["target"][:-20], 
                                     diabetes["target"][-20:])
```

```python
[cosa.shape for cosa in [X_train, X_test, y_train, y_test]]
```

```python
# Forma 2 con train_test_split
# test_size = 422
# shuffle=False
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(diabetes["data"], diabetes["target"],test_size=20, shuffle=False)
```

```python
def predictor_polinomico(ordina=1):
    from sklearn.preprocessing import PolynomialFeatures
    polinomial_transformer = PolynomialFeatures(ordina)
    polinomial_transformer.fit(X_train)
    X_train_transformed = polinomial_transformer.transform(X_train)    

    linear_reg = LinearRegression()
    linear_reg.fit(X_train_transformed, y_train)


    X_test_transformed = polinomial_transformer.transform(X_test)
    predictions = linear_reg.predict(X_test_transformed)

    # Muestra MAE, MAPE, MSE, RMSE
    from sklearn.metrics import mean_absolute_error, mean_absolute_percentage_error ,mean_squared_error, r2_score
    print("MAE", mean_absolute_error(y_test, predictions))
    print("mean_absolute_percentage_error", mean_absolute_percentage_error(y_test, predictions))
    print("mean_squared_error", mean_squared_error(y_test, predictions))
    print("root_mean_squared_error", np.sqrt(mean_squared_error(y_test, predictions)))
    print("r2_score", r2_score(y_test, predictions))
    
predictor_polinomico(1)
```

```python
# Crea una predicción con los datos de test y grábalos en la variable predictions_std
predictions_std = diabetes_model_sc.predict(X_test_sc)
predictions_std
```

```python
# Muestra MAE, MAPE, MSE, RMSE
from sklearn.metrics import mean_absolute_error, mean_absolute_percentage_error ,mean_squared_error
print("MAE", mean_absolute_error(y_test, predictions_std))
print("mean_absolute_percentage_error", mean_absolute_percentage_error(y_test, predictions_std))
print("mean_squared_error", mean_squared_error(y_test, predictions_std))
print("root_mean_squared_error", np.sqrt(mean_squared_error(y_test, predictions_std)))
```
